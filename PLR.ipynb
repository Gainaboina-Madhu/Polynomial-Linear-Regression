```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Position_Salaries.csv')
```

```python
df=df.drop(['Position'], axis = 1)
```

```python
df
```

```python
X=df[['Level']]  #independent
y=df['Salary']   #dependent
```

```python
plt.figure(figsize=(5,3))
plt.title('POLYNOIMAL REGRESSSION')
plt.scatter(x=X,y=y,color='r',marker='*')
plt.show()
```

give X to polynomial linear regression

```python
import sklearn

from sklearn.preprocessing import PolynomialFeatures

poly_reg=PolynomialFeatures(degree=2)
poly_reg.fit(X)
poly_res=poly_reg.transform(X)
poly_res                               #poly_res(X)-> independent
```

```python
#now give to linear regression


from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(poly_res,y)
```

```python
prediction=reg.predict(poly_res)
```

```python
prediction
```

```python
reg.coef_ #mvalue
reg.intercept_  #cvalue
```

```python
plt.figure(figsize=(5,3))
plt.title('POLYNOIMAL REGRESSSION')
plt.scatter(x=X,y=y,color='r',marker='*')
plt.plot(X, prediction, color='b', marker='*')
plt.show()
```
